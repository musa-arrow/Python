```python
class Kumanda():
  def __init__(self,kanal_listesi):
    self.kanal_listesi = kanal_listesi
    self.index = -1
  def __iter__(self):
    return self
  def __next__(self):
    self.index += 1
    if self.index < len(self.kanal_listesi):
      return self.kanal_listesi[self.index]
    else:
      self.index = -1
      raise StopIteration
```

```python
kumanda = Kumanda(["Atv","Trt","Fox","Kanal D","Bloomberg"])
```

```python
iterator = iter(kumanda)
```

```python
next(iterator)
```

```python
def kareleri_al():
  sonuc = []

  for i in range(1,6):
    sonuc.append(i ** 2)
  return sonuc
```

```python
print(kareleri_al())
```

```python
def kareleri_al():
  for i in range(1,6):
    yield i ** 2
```

```python
generator = kareleri_al()
```

```python
iterator = iter(generator)
```

```python
next(iterator)
```

```python
# Generator örnek
```

```python
def carpim_tablosu():
  for i in range(1,11):
    for j in range(1,11):
      yield "{} x {} = {}".format(i,j,i*j)
```

```python
for i in carpim_tablosu(): # for döngüsü zaten iterator mantığıyla çalışır
  print(i)
```
